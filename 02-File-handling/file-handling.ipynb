### File handling in python

Reading a file

```python
with open("sample.txt", "r") as file:
    content = file.read()
    print(content)
```

Writing to a file (overwrites the existing content)

```python
with open("sample.txt", "w") as file:
    file.write("Hello world\n")
    file.write("I am Abhijay Anoop")
```

Writing to a file (without overwriting)

```python
# use append mode
with open("sample.txt", "a") as file:
    file.write("Hello world\n")
    file.write("I am Abhijay Anoop")
```

"w+" mode in python is used to read and write to a file. if the file does not exist it will be created.

```python
# use append mode
with open("example.txt", "w+") as file:
    file.write("Hello world\n")
    file.write("I am Abhijay Anoop")

    file.seek(0)

    content = file.read()
    print(content)
```
