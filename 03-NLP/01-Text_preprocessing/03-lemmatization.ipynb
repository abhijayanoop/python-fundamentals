### Lemmatization

nltk provides WordNetLemmatizer for implementing lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
"""
lemmatizer.lematize(word, pos="n")
pos => postag
it can take any of the following values:
n = noun
v = verb
a = adjective
r = adverb
"""
```

```python
lemmatizer.lemmatize("going", pos="v")
```

```python
words = ['eaten', 'eating', "gone", 'going', 'writing', 'written', 'wrote', 'history']

for word in words:
    print(f"{word} ----> {lemmatizer.lemmatize(word, pos="v")}")
```
