```python
sentence = "The Eiffel Tower is a lattice tower on the Champ de Mars in Paris, France. It is named after the engineer Gustave Eiffel, whose company designed and built the tower from 1887 to 1889."
```

```python
import nltk
```

```python
words = nltk.word_tokenize(sentence)
```

```python
pos_tags = nltk.pos_tag(words)
```
