### Stemming

Stemming is the process of reducing a word to its word step affixes to suffixes and prefixes or to the root of the word called lemma
Important in 
1. Natural Language Understanding (NLU)
1. Natural Language Processing (NLP)

Example: words ['going', 'gone', 'go'] are reduced to 'go'

```python
words = ['eaten', 'eating', "gone", 'going', 'writing', 'written', 'wrote', 'history']
```

```python
import nltk
nltk.download('punkt')
```

1. PorterStemmer

```python
from nltk.stem import PorterStemmer, SnowballStemmer

stemmer = PorterStemmer()
```

```python
for word in words:
    print(f"{word}---->{stemmer.stem(word)}")
```

2. SnowballStemmer

```python
stemmer = SnowballStemmer("english")
```

```python
for word in words:
    print(f"{word}---->{stemmer.stem(word)}")
```
