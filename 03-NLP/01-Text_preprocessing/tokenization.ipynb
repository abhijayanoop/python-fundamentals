```python
corpus = """
Hello my name is Abhijay Anoop. I am a Backend engineer transitioning into AI Engineering. I am very excited.
"""
```

```python
corpus
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
# Converting the corpus (paragraph) into words
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
documents = sent_tokenize(corpus)
for doc in documents:
    print(doc)
```

```python
# Convert a paragraph/sentence into word
words = word_tokenize(corpus)
print(words)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

Using TreebankWordTokenizer

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
